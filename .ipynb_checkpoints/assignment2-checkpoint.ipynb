The following columns were kept for data cleaning purposes:

Not interested in:

```python
stations = pd.read_csv('data\\TfL station data detailed\\Stations.csv')

stations.head()
```

```python
def drop_columns(df, columns):
    """
    Drop specified columns from the DataFrame. Handles the exception when columns aren't found.
    Assumes inplace = True for all operations
    
    Parameters:
    df (DataFrame): The DataFrame from which to drop columns.
    columns (list): List of column names to drop.
    
    Returns:
    DataFrame: The DataFrame with specified columns dropped.
    """
    for col in columns:
        if col in df.columns:
            df.drop(columns=[col], inplace=True)
    return df

columns_to_drop = ['HubNaptanCode', 'BlueBadgeCarParking', 'OutsideStationUniqueId', 'AirportInterchange', 'EmiratesAirLineInterchange']
drop_columns(stations, columns_to_drop)

#Fix NA values
stations['BlueBadgeCarParkSpaces'] = stations['BlueBadgeCarParkSpaces'].replace(np.nan, 0).astype(int)
stations['PierInterchange'] = stations['PierInterchange'].fillna('None')
stations.fillna(False, inplace=True)

#Turn true/false columns into boolean
stations['MainBusInterchange'] = stations['MainBusInterchange'].astype(bool)
stations['NationalRailInterchange'] = stations['NationalRailInterchange'].astype(bool)

stations.head()
```

```python
#Handle the 'FareZones' column

print('Before:', stations['FareZones'].unique())

stations['FareZones'] = stations['FareZones'].replace('Outside', '10')

#Handles tram interchanges. Trams are not included in this analysis
if 'TramInterchange' not in stations.columns:
    stations['TramInterchange'] = stations['FareZones'].str.endswith('|Trams fare zone')
    stations['FareZones'] = stations['FareZones'].replace('Trams fare zone', np.nan)
    stations['FareZones'] = stations['FareZones'].str.replace('|Trams fare zone', '', regex=False)

stations.dropna(inplace=True)

def avg_string(arr):
    """
    Converts an array consisting of string of numbers in the format 'int,int2' etc. to a float by taking its arithmetic mean
    Parameters:
    arr (str): A string of numbers separated by '|'.
    :return: float
    """
    arr = arr.split('|')
    arr = [int(i) for i in arr]
    return np.mean(arr)

#if farezones is a string
stations['FareZones'] = stations['FareZones'].apply(lambda x: avg_string(x) if isinstance(x, str) and '|' in x else x)
stations['FareZones'] = stations['FareZones'].astype(float)

print('After:', stations['FareZones'].unique())

stations[stations['TramInterchange'] == True]
```

```python
def map_boolean(df, to_categorical=True):
    """
    Maps boolean values in a DataFrame column to integers (1 for True, 0 for False).
    
    Parameters:
    df (DataFrame): The DataFrame containing the column to be mapped.
    column (str): The name of the column to map.
    
    Returns:
    DataFrame: The DataFrame with the specified column mapped to integers.
    """
    for col in df.columns:
        if df[col].dtype == 'bool':
            df[col] = df[col].map({False: 0, True: 1})
            if (to_categorical):
                df[col] = pd.Categorical(df[col], categories=[0, 1], ordered=True)
    return df

stations = map_boolean(stations)

if stations['PierInterchange'].dtype == 'O':
     stations['PierInterchange'] = stations['PierInterchange'].map({'None': 0, 'Partial': 1, 'Full': 2})
     stations['PierInterchange'] = pd.Categorical(stations['PierInterchange'], categories=[0, 1, 2], ordered=True)

stations
```

### 3.3 StationPoints dataset

Data that includes the geographical location of each station. The Latitude and Longitude will be used as predictors for the model.

Multiple latitude and longitude data has been provided for each station (i.e. different latitude and longitude data for footbridge, platforms, councourse etc.). The first latitude and longitude provided for each station will be used. It shouldn't make a big difference even if the latitude and longitude combination is describing different part of the stations (i.e. the combination used for London Bridge station may be the concourse location, and the combination used for Liverpool Street is the platform location) as stations are assumed to be connected and not far apart from each other. 

```python
stationpoints = pd.read_csv('data\\TfL station data detailed\\StationPoints.csv')

stationpoints = stationpoints[['StationUniqueId', 'Lat', 'Lon']]
stationpoints.rename(columns = {'StationUniqueId': 'UniqueId', 'Lat': 'Latitude', 'Lon':'Longitude'}, inplace=True)

#Drop all duplicates and use only the first latitude and longitude
stationpoints = stationpoints.drop_duplicates(subset=['UniqueId'])

drop_columns(stations, ['Latitude', 'Longitude'])

stations = stations.merge(stationpoints, how='left', on='UniqueId')
stations
```

### 3.4 PlatformServices dataset

Contains data describing services on each platform.

It is used to extract the number of lines (excluding National Rail, ThamesLink and Trams) the station serves. The number of lines will also be used as a predictor.

The `NationalRailInterchange` column in the Stations dataset will also be modified, as some stations that serve both National Rail (including ThamesLink) and TfL operated services were not identified in the original column, but listed here.

```python
platforms = pd.read_csv('data\\TfL station data detailed\\PlatformServices.csv')
platforms = platforms[['PlatformUniqueId', 'Line']]
platforms.rename(columns = {'PlatformUniqueId': 'UniqueId'}, inplace=True)

platforms['UniqueId'] = platforms['UniqueId'].str.split('-').str[0]

platforms['NationalRailInterchange'] = platforms['Line'].str.contains('national-rail')
map_boolean(platforms)

# #Drop tram and cable car lines
platforms = platforms[~platforms['Line'].str.contains('tram')]
platforms = platforms[~platforms['Line'].str.contains('london-cable-car')]

platforms = platforms.drop_duplicates()

# Handle national rail interchanges
platforms = platforms.groupby('UniqueId').agg({'Line': 'count', 'NationalRailInterchange': 'max'}).reset_index()
platforms.rename(columns = {'Line': 'NumLines'}, inplace=True)
#check each row whether national rail interchange = 1. if so, number of lines -= 1
platforms['NumLines'] = np.where(platforms['NationalRailInterchange'] == 1, platforms['NumLines'] - 1, platforms['NumLines'])

if 'NumLines' not in stations.columns:
    stations = stations.merge(platforms, how='left', on='UniqueId')
    stations['NationalRailInterchange_x'] = stations['NationalRailInterchange_x'].astype(bool)
    stations['NationalRailInterchange_y'] = stations['NationalRailInterchange_y'].astype(bool)
    stations['NationalRailInterchange'] = stations['NationalRailInterchange_x'] | stations['NationalRailInterchange_y']
    drop_columns(stations, ['NationalRailInterchange_x', 'NationalRailInterchange_y'])

map_boolean(stations)

stations
```

### 3.5 Station Footfall dataset

Contains data of the number of people entering and exiting the station. Columns include:

- `TravelDate`: Date of travel
- `DayOfWeek`: String, containing the day of week
- `EntryTapCount`: Number of 'tap ons' on the day
-` ExitTapCount`: Number of 'tap offs' on the day

To eliminate the time-series nature of the data, entry and exit counts for each station will be grouped by the DayOfWeek.

The following cell includes adapted code originally written by jonas & Mortensen (2022).

```python
#Load entry and exit data

import glob

#Combine all footfall data into one dataframe
path = 'data/station_footfall'
fnames = glob.glob(path + '/*.csv')
footfall = []
for fname in fnames:
    footfall.append(pd.read_csv(fname, parse_dates=[0]))
footfall[0] = footfall[0].rename(columns={'DayOFWeek': 'DayOfWeek'})
footfall = pd.concat(footfall, ignore_index=True).drop_duplicates().reset_index(drop=True)

footfall.head()
```

```python
# Group station by DayOfWeek
footfall = footfall.groupby(['Station', 'DayOfWeek']).agg({'EntryTapCount': 'sum',
                                               'ExitTapCount': 'sum'}).reset_index()

# Map the day of the week to a number
if footfall['DayOfWeek'].dtype == 'O':
    footfall['DayOfWeek'] = footfall['DayOfWeek'].map({'Monday': 1, 'Tuesday': 2, 'Wednesday': 3, 'Thursday': 4, 'Friday': 5, 'Saturday': 6, 'Sunday': 0})

# Set day of week to categorical
footfall['DayOfWeek'] = pd.Categorical(footfall['DayOfWeek'], categories=[0, 1, 2, 3, 4, 5, 6], ordered=True)

footfall
```

```python
#Remove DLR, NR, ... etc. from the station names
footfall['Station'] = footfall['Station'].str.replace(' DLR', '', regex=False)
footfall['Station'] = footfall['Station'].str.replace(' NR', '', regex=False)
footfall['Station'] = footfall['Station'].str.replace(' EL', '', regex=False)
footfall['Station'] = footfall['Station'].str.replace(' LO', '', regex=False)
footfall['Station'] = footfall['Station'].str.replace(' SR', '', regex=False)
footfall['Station'] = footfall['Station'].str.replace(' C&H', '', regex=False)
footfall['Station'] = footfall['Station'].str.replace(' D&P', '', regex=False)

#Handle Heathrow stations
footfall['Station'] = footfall['Station'].str.replace(' TfL Rail/HEx', '', regex=False)
footfall['Station'] = footfall['Station'].str.replace('T2&3', 'Terminals 2 & 3', regex=False)
footfall['Station'] = footfall['Station'].str.replace('T4', 'Terminal 4', regex=False)
footfall['Station'] = footfall['Station'].str.replace('T5', 'Terminal 5', regex=False)
footfall['Station'] = footfall['Station'].str.replace('2&3', '2 & 3', regex=False)

#First line to handle stations named 'Liverpool St'
footfall['Station'] = footfall['Station'].str.replace('Liverpool Street', 'Liverpool St', regex=False)
footfall['Station'] = footfall['Station'].str.replace('Liverpool St', 'Liverpool Street', regex=False)

#Combine stations with same station names
footfall = footfall.groupby(['Station', 'DayOfWeek']).agg({'EntryTapCount': 'sum', 'ExitTapCount': 'sum'}).reset_index()

footfall[footfall['Station'].str.contains('Liverpool')]
```

## 5. Model Selection

To use linear regression models, the multicollinearity issue for `Latitude` must be dealt with. `Latitude` must be dropped as a predictor, or other transformations must be done; yet, it is not possible to `Latitude` by itself without dropping `Longitude`, as both variables together represent a single geographic point. An alternative is to combine these two variables to a separate category using clustering algorithms, yet there's no guarantee even after clustering the multicollinearity issue will be solved. 

Additionally, the default way `scikit-learn` handles categorical variable predictors by converting them to numerical variables (unlike R), which will lead to incorrect model equations and assumptions. Therefore, to use parametric regression models, much more data cleaning and transformation must be done, and there isn't a need to do so considering other models exist.

The independence assumption for Naive Bayes is violated considering there is significant multicollinearity within the data, therefore it will not be considered.

For K-Nearest Neighours (KNN), there is no underlying assumption that the data must follow a specific distribution, which suits the data well. Yet, the algorithm does not perform well with data containing outliers. This could be potentially fixed by scaling, decreasing the Eucliean distance between each datapoint (Baker 2025).

Most predictors in this model are categorical variables, which makes tree based regressors ideal. The problem where `scikit-learn` converts categorical variables to numerical variables no longer matters (i.e. say variable $X$ contains two possible values encoded as 0 and 1, if data is split on $X <= 0.5$, it is equivalent to splitting on $X=0$).

Neural Networks are not considered due to the heavy computation required to train the models. The data contains only approximately 2000-3000 rows and the project is relatively small-scaled. Tree based regressors should be sufficient to capture the relationship between the predictors.

## 6. Approach

**EntryTapCount prediction**:

The goal is to develop a single model that best predict `EntryTapCount`.

The data will be splitted into train (75%), validation (12.5%) and test sets (12.5%). If required, scaling will be done on the data.

KNN, Decision Tree and Random Forest regressors will all be potential candidates. Prior to training the models, cross validation will be used to compare model performance before and after scaling the dataset. Then, optimal hyperparameters for each model (after scaling if required) will be tuned using `GridsearchCV`, and the best estimator found will be used for comparison on the validation set. A Dummy Regressor will also be included to as a base model.

The best performing model will then be chosen, and tested using the test set. The final metrics will be the model accuracy.

*Note: Since Decision Trees performed very well during the initial model validation, Random Forest was then added as a candidate model despite the heavy computation required (the same reason which led me to not use Neural Networks) to see if an ensemble model involving multiple trees can outperform a single tree.*

**ExitTapCount prediction**:

This approach takes advantage of the strong linear relationship between `EntryTapCount` and `ExitTapCount`.

The train and test sets for `ExitTapCount` will contain the same rows as the train and test sets used to predict `EntryTapCount`. The only exception is that model validation will be conducted using cross validation, hence the validation set is combined with the training set. 

Ordinary Least Squares, Theil-Sen, RANSAC and Huber regressors will be training using the true values of `EntryTapCount`. The best performing linear regression model will be used to model `ExitTapCount` using the predicted values from the `EntryTapCount` model. Model will then be evaluated.

**Model evaluation**:

Three metrics will be used:

- **Root Mean Squared Error** (RMSE): Main metric used for model comparison (Ogozaly 2021). Gives an insight on the prediction accuracy, excluding noise (Ogozaly 2021). Lower the RMSE value, the greater the model accuracy.
- **Median Absolute Error**: A more robust metric that takes account of outliers. Considering that the data is positively skewed, this metric gives an picture of how well the model can predict footfall figures for the majority of stations.
- **R-squared** ($R^2$): An $R^2$ score of 51% implies 51% of the variance within the data can be explained by the model. Reports how well the model can handle noise (Ogozaly 2021).

## 7. Predicting EntryTapCount


### 7.1 K-Nearest Neighbours

KNN is a lazy learning algorithm that makes predictions by aggregating the target value of the most similar $k$ observations (determined by the distance between the new point and each datapoint) in the feature space (Baker 2025; GeeksforGeeks 2024a). Several advantages and disadvantages of KNN regressors suggested by Baker (2025) and Pedamkar (2023) include:

✅ Easy to implement. Very little hyperparamaters are required to be tuned during model training.

✅ Non-linear, non-parametric regressor. Does not assume data follows any specific distribution.

✅ Suitable for both numerical and categorical data, without the need for extensive data cleaning.

✖️ Outliers significantly decrease model performance well despite it being a non-parametric regressor, as the distance of outliers to other points are much larger in comparison. They must be removed prior to model fitting. Scaling the data may also fix this issue.

✖️ As dimensionality increases, the prediction accuracy is likely to decrease (more likely if predictors are irrelevant)

✖️ Computationally expensive, especially for large datasets, as all training data is required to be used during prediction.

The approach to train the KNN model is referenced from MH Yu (2025).

```python
#import train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
X_valid, X_test, y_valid, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=42)

X_train.shape, X_valid.shape, X_test.shape
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

n_neighbors = np.arange(1, 11, 1)
scalers = ['Unscaled', StandardScaler(), MinMaxScaler(), RobustScaler()]
scaler_names = ['Unscaled', 'StandardScaler', 'MinMaxScaler', 'RobustScaler']

cv_scores = []
cv_scores_std = []
for s in scalers:
    if s == 'Unscaled':
        X_train_scaled = X_train
    else:
        X_train_scaled = s.fit_transform(X_train)

    cv_scores_temp = []
    cv_scores_std_temp = []
    for k in n_neighbors:
        knn = KNeighborsRegressor(n_neighbors=k)
        scores = cross_val_score(knn, X_train_scaled, y_train, cv=10, scoring='neg_root_mean_squared_error')
        cv_scores_temp.append(np.mean(scores))
        cv_scores_std_temp.append(np.std(scores))
    
    cv_scores.append(cv_scores_temp)
    cv_scores_std.append(cv_scores_std_temp)

for i in range(len(scalers)):
    sns.lineplot(x=n_neighbors, y=cv_scores[i], label=scaler_names[i], marker='o')

plt.title('KNN Cross-Validation Scores')
plt.xticks(n_neighbors)
plt.xlabel('Number of Neighbors')
plt.ylabel('Negative RMSE')
plt.show()
```

Expectedly, using `RobustScaler()` to scale the data leads to the lowest RMSE on average, since it scales the data using the IQR and median, handling the skewness of the data used. `RoubstScaler()` will be used to scale data prior to fitting KNN.

```python
from sklearn.model_selection import GridSearchCV

scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train)

knn = KNeighborsRegressor()
param_grid = {
    'n_neighbors': np.arange(1, 11, 1),
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
}

def to_hr_min_sec(seconds):
    """
    Convert seconds to a string in the format 'HH:MM:SS'.
    
    Parameters:
    seconds (int): Number of seconds.
    
    Returns:
    str: Time in 'HH:MM:SS' format.
    """
    hr = seconds / 3600
    min = (seconds % 3600) / 60
    sec = seconds % 60
    return f"{int(hr):02}:{int(min):02}:{int(sec):02}"

start = time.time()
grid_knn = GridSearchCV(knn, param_grid, cv=10, scoring='neg_root_mean_squared_error', n_jobs=-1)
grid_knn.fit(X_train_scaled, y_train)
end = time.time()
print('Training time:', to_hr_min_sec(end - start))
print('Best parameters:', grid_knn.best_params_)
print('Best score:', grid_knn.best_score_)
```

```python
from sklearn.metrics import root_mean_squared_error, median_absolute_error, r2_score

X_valid_scaled = scaler.transform(X_valid)
y_pred_knn = grid_knn.predict(X_valid_scaled)

def get_metrics(y_true, y_pred, time, ret=True):
    """
    Calculate and print regression metrics.
    
    Parameters:
    y_true (array-like): True target values.
    y_pred (array-like): Predicted target values.
    time (float): Time taken for the prediction.
    ret (bool): Default = True.

    Returns:
    list: Metrics of model as a list if ret is True, otherwise None.
    """
    rmse = root_mean_squared_error(y_true, y_pred)
    mae = median_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    
    print('RMSE: %.4f' %rmse)
    print('Median Absolute Error: %.4f' %mae)
    print('R2: %.4f' %r2)
    
    if ret:
        return [rmse, mae, r2, time]

results = pd.DataFrame(columns=['RMSE', 'Median Absolute Error', 'R2', 'Training Time'])
results.loc['KNN'] = get_metrics(y_valid, y_pred_knn, end-start)
```

```python
def plot_results(y_true, y_pred, title):
    """
    Cretes a scatterplot model predictions using plotly express.
    
    Parameters:
    y_true (Series): True values.
    y_pred (Series): Predicted values.
    title (str): Title for the plot.
    """
    fig = px.scatter(x=y_true, y=y_pred, color=abs(y_true - y_pred), hover_name=y_true.index)
    fig.update_traces(marker=dict(size=5, opacity=0.5, line=dict(width=1, color='DarkSlateGrey')),
                    selector=dict(mode='markers'))
    fig.update_layout(title=title, xaxis_title='True Values', yaxis_title='Predicted Values', coloraxis_colorbar=dict(title='Error'))
    fig.show()

plot_results(y_valid, y_pred_knn, 'KNN Model Predictions')
```

### 7.2 Decision Tree

Decision Tree is a classification/regression algorithm that makes predictions by recursively/iteratively (not sure how it's implemented in `sklearn`) splitting the data based on their features (Zhou et al. 2025). Several advantages and disadvantages of Decision Trees regressors suggested by GeeksforGeeks (2024b) and Zhou et al. (2025) include:

✅ Easy to interpret (compared to algorithms such as neural networks).

✅ Non-parametric model. Does not assume data follows any specific distribution. This also implies that the algorithm can effectively handle outliers and skewed data.

✅ Suitable for both numerical and categorical data, without the need for extensive data cleaning.

✖️ Overfitting. As the tree gets more complex, the more likely the tree will capture the random noise of the data. This can be handled by setting limits, e.g. setting a maximum depth of the tree or setting the maximum number of leaf nodes.

✖️ Loses information when dealing with continuous numerical variables.

```python
from sklearn.tree import DecisionTreeRegressor

cv_scores = []
cv_scores_mean = []
cv_scores_std = []
for s in scalers:
    if s == 'Unscaled':
        X_train_scaled = X_train
    else:
        X_train_scaled = s.fit_transform(X_train)

    dt = DecisionTreeRegressor(random_state=42)
    scores = cross_val_score(dt, X_train_scaled, y_train, cv=10, scoring='neg_root_mean_squared_error')
    cv_scores.append(scores)
    cv_scores_mean.append(np.mean(scores))
    cv_scores_std.append(np.std(scores))

fig = px.bar(x=scaler_names, y=cv_scores_mean, error_y=cv_scores_std, title='Decision Tree Regressor Scores')
fig.update_layout(xaxis_title='Scaler', yaxis_title='Negative RMSE', title_x=0.5)
fig.show()
```

No significant difference in model performance when using scalers. Will use `RobustScaler()` to ensure consistency with KNN model.

```python
dt = DecisionTreeRegressor(random_state=42)

param_grid = {
    'max_depth': np.arange(1, 41, 1),
    'min_samples_split': np.arange(2, 6, 1),
    'min_samples_leaf': np.arange(1, 6, 1)
}

start = time.time()
grid_dt = GridSearchCV(dt, param_grid, cv=10, scoring='neg_root_mean_squared_error', n_jobs=-1)
grid_dt.fit(X_train_scaled, y_train)
end = time.time()
print('Training time:', to_hr_min_sec(end - start))
print('Best parameters:', grid_dt.best_params_)
print('Best score:', grid_dt.best_score_)
```

```python
dt = grid_dt.best_estimator_

dt.fit(X_train_scaled, y_train)
y_pred_dt = dt.predict(X_valid_scaled)

def append_result(model_name, y_pred, time):
    """
    Append the model name and its metrics to the results DataFrame.
    Also prints the results of the model.
    """
    drop_columns(results, ['Model'])
    results.loc[model_name] = get_metrics(y_valid, y_pred, time)

append_result('Decision Tree', y_pred_dt, end-start)
```

### 7.3 Random Forest

Random Forest is an ensemble method that aggregates the predictions of several decision trees to generate a final prediction. Several advantages and disadvantages of Random Forest regressors suggested by GeeksforGeeks (2024b) include:

✅ Averages the results of several decision trees which is likely to lead to increased prediction accuracy.

✅ Lowers the probability of noise within the data affecting the prediction of the model as multiple trees have to be combined.

✅ Every other advantage that decision trees have.

✖️ High computation complexity, and significant time required to train the model as number of estimators increase due to the need to combine variuos trees together for prediction.

✖️ Every other disadvantage that decision trees have.

```python
from sklearn.ensemble import RandomForestRegressor

cv_scores = []
cv_scores_mean = []
cv_scores_std = []
for s in scalers:
    if s == 'Unscaled':
        X_train_scaled = X_train
    else:
        X_train_scaled = s.fit_transform(X_train)

    rf = RandomForestRegressor(random_state=42)
    scores = cross_val_score(rf, X_train_scaled, y_train, cv=10, scoring='neg_root_mean_squared_error')
    cv_scores.append(scores)
    cv_scores_mean.append(np.mean(scores))
    cv_scores_std.append(np.std(scores))

fig = px.bar(x=scaler_names, y=cv_scores_mean, error_y=cv_scores_std, title='Random Forest Regressor Scores')
fig.update_layout(xaxis_title='Scaler', yaxis_title='Negative RMSE', title_x=0.5)
fig.show()
```

Again, no significant difference when scaling data. Will continue to use `RobustScaler()`.

Note: The time to run GridSearch is long. Set `run_cell` to `False` to skip the running of the cell when running all.

Cell below includes adapted code originally by Beheshti (2022).

```python
run_cell = True

grid_rf = None   

if run_cell:
    start = time.time()
    rf = RandomForestRegressor(random_state=42)
    # Note: the param_grid has been tuned based on previous runs. Original ranges were much larger.
    param_grid = {
        'n_estimators': np.arange(80, 101, 1),
        'max_depth': np.arange(20, 31, 1),
        'min_samples_split': np.arange(2, 5, 1),
        'min_samples_leaf': np.arange(1, 4, 1)
    }
    grid_rf = GridSearchCV(rf, param_grid, cv=10, scoring='neg_root_mean_squared_error', n_jobs=-1, verbose=1)
    grid_rf.fit(X_train_scaled, y_train)
    end = time.time()
    print('Training time:', to_hr_min_sec(end - start))
    print('Best parameters:', grid_rf.best_params_)
    print('Best score:', grid_rf.best_score_)
```

## 10. Model Reflection

### 10.1 Why was the model this accurate?

The normalised RMSE and Median Absolute Error is approximately is $0.0172$ and $0.058$ for both models, very close to zero, suggesting that the model has high perdiction accuracy (Bobbitt 2021). The $R^2$ value is also above 95% which implies the model can explain close to all of the variability of the data, and the residual values are really small. In other words, the model is pretty much spot on predicting the tap counts of a station.

The reason why the model performed so well can be either:

- It's just a very good model.
- **There are too many relevant predictors and information is leaking between the target and predictor variables** (Baker, personal communication, 26 May 2025).

In this case, the latter is more likely than the prior. The model includes predictors that infer from the target variables. For example, TfL is more likely to install Wifi, toilets and other facilities based on the number of passengers already using, or predicted to use the station. Therefore, `HasWifi` and `HasToilets` convey information that is somewhat dependent on the independent variable (Baker, personal communication, 26 May 2025). The same reason makes this model **impractical to use when building new stations, deviating from the original goal**.

### 10.2 What could be improved?

Unfortunately at the time of writing it's too late to add these improvements. But here's what could've been done if the model was to be rebuilt again:

- **Improving the model comparison methodology**. Only one validation set was used to compare the different models, and this caused an issue due to the performance of Decision Trees and Random Forest being very close. As there was only one score per metric to compare on, no hypothesis test could be conducted to test for significance. Random Forest may well have performed worse on a different validation set. A 3-5 fold cross validation could have been done to produce a more accurate result.

- **Change the predictors**, so that the model actually aligns with the original purpose. This can be improved by:
    - **Eliminating the predictors that are dependent on the target variable**. Frankly speaking, only `DayOfWeek`, `Latitude`, `Longitude` and `NumLines` in this model are predictors that would be intially considered when building a new station.
    - **Adding new predictors that are relevant and practical**. For example, network analysis can be conducted prior to model fitting, since all train stations are either already part of, or planned to be incorporated in the rail network, and eigenvector centrality can be calculated and used as a predictor.
    - **Considering training time vs performance** (Beheshti 2022). The original decision to include Random Forest is to see if the model can outperform Decision Trees, which was a good fit on the validation data. This hypothesis ended up to be true, yet the training time for Random Forest was over 100 times longer in comparison (excluding previous runs). For a small-scaled project like this one, the only that is sacrificed is the computer's RAM and my time doing other assignments. However, for larger-scaled project, this may lead to increased labour and opportunity costs. There's no definite answer on whether it is worth spending substantially more time to train a model that performs slightly better; but in this case, where the evidence of Random Forest performing better than Decision Trees is relatively weak, this decision is probably not justified.

### 10.3 How can this model (still) be used?

There is still practical uses for this model, in cases where:

- When TfL suddenly decides to not release station footfall figures. This could happen during various crises, e.g. wars, times of political tension.
- To predict station patronage for current stations when modifying stations. e.g. If TfL decides to expand Liverpool Street Station and increase the number of lines going through, it can use it to project new station patronage. Since Wifi facilities has already been installed at the station, `HasWifi` will be a valid predictor.
